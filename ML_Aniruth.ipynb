```python
# 7:
# I am getting the movie which has highest number of ratings. I did it using value_counts method and then got the highest count 
# using idxmax method.
highest_rating = df_ratings['movie_id'].value_counts()
highest = highest_rating.idxmax()
print("\nMovie with highest ratings:", highest)
```

```python
# 8:
# Here, I am getting the movie that got the highest 5 star ratings. Finally, counting the highest ratings and the movie ID.
rating_5_movie = df_ratings[df_ratings['rating'] == 5]['movie_id'].value_counts().idxmax()
print("The movie that received the highest '5' rating is Movie ID:", rating_5_movie)
```

```python
# 9:
# Plotted histogram using hist() method on the ocuurence of user id in the df_ratings dataframe.
rating_count_user_id = df_ratings['user_id'].value_counts()
rating_count_user_id.hist(bins=30)
plt.title("Ratings by User")
plt.xlabel("Number of Ratings")
plt.ylabel("No. of occurences")
plt.show()
```

```python
# 10:
# Calculating and printing the year where most number of movies have been released.
# For this, I have converted the 'release_date' column to a datetime format and got the frequency of each release year.
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], errors='coerce')
most_releases_year = df_movies['release_date'].dt.year.value_counts().idxmax()
print("\nYear with the greatest number of movies released is:", most_releases_year)
```

```python
# 11:
# Merged ratings and movies dataframe by column named movie_id
combined_data_frame = pd.merge(df_ratings, df_movies, on='movie_id')
print("\nSample rows from the combined one:")
print(combined_data_frame.sample(10))
```

```python
# 12:
# Filtering Star Wars movie and the getting the length of the ratings.
movie_ratings = combined_data_frame[combined_data_frame['title'] == 'Star Wars (1977)']
num_movie_ratings = len(movie_ratings)
print("\n'Star Wars (1977)' was rated", num_movie_ratings, "times.")
```

```python
# 13:
# Plotted the histogram for number of ratings per movie in the merged dataframe.
rating_counts_movie = combined_data_frame['movie_id'].value_counts()
rating_counts_movie.hist(bins=50)
plt.title("Movie wise ratings")
plt.xlabel("No. of Ratings")
plt.ylabel("No. of times")
plt.show()
```

```python
# 14:
# Plotting the bar graph by grouping the rating and title. Then I sorted the values in descending order. Printing
# the top 10.
top_movies = combined_data_frame.groupby('title')['rating'].mean().sort_values(ascending=False).head(10)
top_movies.plot(kind='barh')
plt.title("Top Rated Movies")
plt.xlabel("Average Rating")
plt.ylabel("Movie Title")
plt.gca().invert_yaxis()  
plt.show()
```

```python
# 15:
# Based on user_od from df_users dataframe, I combined users to the befper combined dataframe.
df_combined_users = pd.merge(combined_data_frame, df_users, on='user_id')
```

```python
# 16
# Printing number of ratings and the average of rating for each movie. I use agg() to do that after grouping them by rating and 
# title.
movie_ratings_stats = df_combined_users.groupby('title')['rating'].agg(['count', 'mean'])
print("\nNumber of ratings, average rating for each movie:")
print(movie_ratings_stats)
```

```python
# 17:
# With a minimum rating of 100, rating the top arted movies.
top_rated_100_min = movie_ratings_stats[movie_ratings_stats['count'] >= 100].sort_values(by='mean', ascending=False)
print("\nTop-rated movies with a minimum rating count of 100:")
print(top_rated_100_min)
```

```python
# 18:
# Created a histogram based on gender. Then, I showed the averge rating per movie by gender.
rating_gender = df_combined_users.groupby(['title', 'gender'])['rating'].mean().unstack()
rating_gender.plot(kind='hist', alpha=0.5, bins=20, title="Average Rating based on Gender")
plt.xlabel("Average Rating")
plt.ylabel("Frequency")
plt.legend(title="Gender", labels=["Male", "Female"])
plt.show()
```
